```python
years = range(2010, 2021)
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(consumption_oil, years, 'tab:blue')
axs[0, 0].set_title('1')
axs[0, 1].plot(consumption_gas, years, 'tab:red')
axs[0, 1].set_title('2')
axs[1, 0].scatter(consumption_renewable, years, c='yellow')
axs[1, 0].set_title('3')
axs[1, 1].scatter(consumption_oil, years, c='green')
axs[1, 1].set_title('4')
```
